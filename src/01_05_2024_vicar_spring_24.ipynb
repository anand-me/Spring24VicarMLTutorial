 # *Let's work on **lists**

```python
# Creating the List
members_of_CTML = ['Akshay', 'Alireza', 'Gautam', 'Aojia', 'Brian', 'Shirin', 'Shahriar', 'KSH']
# Accessing Elements
print("First member:", members_of_CTML[0])
print("Second member:", members_of_CTML[1])
```

```python
# Modifying Elements
members_of_CTML[7] = 'Jason'
print("Updated list:", members_of_CTML)
```

```python
# Slicing the List
sliced_list = members_of_CTML[2:5]
print("Sliced list:", sliced_list)
```

```python
# Removing Elements
removed_member = members_of_CTML.pop(6)
print("Removed member at index 3:", removed_member)
print("Updated list after pop:", members_of_CTML)
```

```python
# Checking if an Element is in the List
if 'Brian' in members_of_CTML:
    print("Yes, 'Brian' is in the list.")
else:
    print("No, 'Brian' is not in the list.")
```

# *Now is the time to play with **Tuples**

```python
# Creating the Tuple
members_of_CTML_tuple = ('Akshay', 'Alireza', 'Gautam', 'Aojia', 'Brian', 'Shirin', 'Shahriar', 'KSH')
# Accessing Elements
print("First member:", members_of_CTML_tuple[0])
print("Second member:", members_of_CTML_tuple[1])
```

# *Note: Since tuples are immutable, modifying and adding elements directly is not possible.*


```python
# Slicing the Tuple
sliced_tuple = members_of_CTML_tuple[2:5]
print("Sliced tuple:", sliced_tuple)
```

```python
# Iterating Through the Tuple
print("Iterating through the tuple:")
for member in members_of_CTML_tuple:
    print(member)
```

```python
# Checking if an Element is in the Tuple
if 'Brian' in members_of_CTML_tuple:
    print("Yes, 'Brian' is in the tuple.")
else:
    print("No, 'Brian' is not in the tuple.")

# Finding the Length of the Tuple
tuple_length = len(members_of_CTML_tuple)
print("Length of the tuple:", tuple_length)
```

 # *Now, is the time to look at **dictionary**

```python
# Creating the Dictionary with Roles (I guess don't get mad or excited this is hypothetical roles :)
members_of_CTML_dict = {
    'Akshay': 'Data Scientist',
    'Alireza': 'Machine Learning Engineer',
    'Gautam': 'CFD Engineer',
    'Aojia': 'Software Developer',
    'Brian': 'Robotics Engineer',
    'Shirin': 'AI Engineer',
    'Shahriar': 'CFD Developer',
    'KSH': 'Scientist'
}
```

```python
# Modifying Elements: Alireza is promoted to AI Engineer (lots of work on unsupervised ML and DL :)
members_of_CTML_dict['Alireza'] = 'AI Researcher'
print("Updated dictionary:", members_of_CTML_dict)
```

```python
# Adding Elements : Potentially PostDoc
members_of_CTML_dict['NewMember'] = 'Software Engineer'
print("Dictionary after adding 'NewMember':", members_of_CTML_dict)
```

```python
# Removing Elements (Rmoved Akshay)
removed_member_role = members_of_CTML_dict.pop('Akshay')
print("Removed role of Akshay:", removed_member_role)
print("Updated dictionary after pop:", members_of_CTML_dict)
```

```python
# Iterating Through the Dictionary
print("Iterating through the dictionary:")
for member, role in members_of_CTML_dict.items():
    print(f"{member}: {role}")
```

```python
# Checking if a Key is in the Dictionary
if 'Shirin' in members_of_CTML_dict:
    print("Yes, 'Shirin' is in the dictionary.")
else:
    print("No, 'Shirin' is not in the dictionary.")
```

# *Simple **algebraic operations using intergers and floats**


```python
# Integer and Floating-point Operations
integer_result = 5 + 3     # Addition
float_result = 5.0 - 2.5   # Subtraction
mixed_result = 4 * 1.5     # Multiplication
# Division
integer_division = 7 / 2  # Integer division
float_division = 7 / 2.0   # Floating-point division
# Exponentiation
power_result = 2 ** 3      # 2 raised to the power of 3
# Modulo (Remainder)
remainder = 10 % 3
```

```python
# Printing Results
print("Integer Result:", integer_result)
print("Float Result:", float_result)
print("Mixed Result:", mixed_result)
print("Integer Division:", integer_division)
print("Float Division:", float_division)
print("Power Result:", power_result)
print("Remainder:", remainder)
```

```python
# numpy library: this is a sophisticated way to wor with arrays in python
import numpy as np
# Creating NumPy Arrays
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([2, 4, 6, 8, 10])
```

```python
# Basic Operations
sum_array = array1 + array2
difference_array = array2 - array1
product_array = array1 * array2
quotient_array = array2 / array1

# Broadcasting: Operations on entire arrays
scaled_array = array1 * 2

# Universal Functions (ufunc)
square_root_array = np.sqrt(array1)
exponential_array = np.exp(array1)
sin_array = np.sin(array1)

# Dot Product
dot_product = np.dot(array1, array2)

# Printing Results
print("Array 1:", array1)
print("Array 2:", array2)
print("Sum:", sum_array)
print("Difference:", difference_array)
print("Product:", product_array)
print("Quotient:", quotient_array)
print("Scaled Array:", scaled_array)
print("Square Root:", square_root_array)
print("Exponential:", exponential_array)
print("Sin:", sin_array)
print("Dot Product:", dot_product)
```
